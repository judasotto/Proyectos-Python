```python
import pandas as pd

# Cargar el dataset
df = pd.read_csv('diamonds.csv')
print(df.head())
```

```python
# Mostrar las primeras filas del dataframe
print(df.head())

# Obtener información general del dataframe
print(df.info())

# Resumen estadístico de las variables numéricas
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribución del precio
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribución del Precio de los Diamantes')
plt.xlabel('Precio (USD)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Relación entre carat y precio
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Relación entre Carat y Precio')
plt.xlabel('Carat')
plt.ylabel('Precio (USD)')
plt.show()
```

```python
# Distribución del corte
plt.figure(figsize=(10, 6))
sns.countplot(x='cut', data=df, order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])
plt.title('Distribución de la Calidad del Corte')
plt.xlabel('Calidad del Corte')
plt.ylabel('Cantidad de Diamantes')
plt.show()
```

```python
# Relación entre clarity y precio
plt.figure(figsize=(10, 6))
sns.boxplot(x='clarity', y='price', data=df)
plt.title('Relación entre Clarity y Precio')
plt.xlabel('Clarity')
plt.ylabel('Precio (USD)')
plt.show()
```

```python
# Codificar las variables categóricas usando one-hot encoding
encoded_df = pd.get_dummies(df)

# Mapa de calor de la correlación entre todas las variables
plt.figure(figsize=(15, 10))
sns.heatmap(encoded_df.corr(), annot=False, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación de Todas las Variables (Codificadas)')
plt.show()
```
